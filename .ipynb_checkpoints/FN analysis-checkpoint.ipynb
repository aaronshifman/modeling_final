```julia
using PyPlot
```

Setup parameters

```julia
b = 0.3;
a = 0.5;
z = -0.001;
dt = 0.001;
tmax = 200;
times = collect(0:dt:tmax);
xs = zeros(length(times));
ys = zeros(length(times));
x_0 = 0; y_0 = 0;
```

Setup differential equations

```julia
x_dot(c,z,x,y) = y+x-(x^3)/3+z;
y_dot(a,b,c,x,y)= -x-a-b*y;
```

Time dependant solver solver

```julia
for (ix,t) in enumerate(times[1:end-1])
    xs[ix+1] = xs[ix]+dt*x_dot(c,z,xs[ix],ys[ix])
    ys[ix+1] = ys[ix]+dt*y_dot(a,b,c,xs[ix],ys[ix])
end
```

Show results

```julia
plot(times,xs)
plot(times,ys)
```
