```julia
using PyPlot
```

We have a 2d system:
$\frac{d\chi}{d\tau} = \gamma + \chi - \frac{\chi^3}{3} + z\\
\frac{d\gamma}{d\tau} = -\chi + \alpha - \gamma\beta
$

```julia
x_dot(x,y,z) = y+x-(x^3)/3+z;
y_dot(x,y,a,b)= -x-a-b*y;
```

```julia
x_nullcline(x,z) = x.^3/3-x-z;
y_nullcline(x,a,b) = (x+a)/b;
```

Setup sample paramaters

```julia
a = 0.5;
b = 0.5;
z = -0.001;
```

$\frac{d\chi}{d\tau} = 0 \Rightarrow \gamma = \frac{\chi^3}{3} -\chi -z \\
\frac{d\gamma}{d\tau} = 0 \Rightarrow \gamma = \frac{\chi+\alpha}{\beta} 
$

```julia
x = [-3:0.01:3];
x_null = x_nullcline(x,z);
y_null = y_nullcline(x,a,b);
plot(x,x_null)
plot(x,y_null)


a = 0.5;
b = 0.3;
z = -0.001;

dt = 0.001;
tmax = 200;
times = collect(0:dt:tmax);
xs = zeros(length(times));
ys = zeros(length(times));
x_0 = 0; y_0 = 0;

for (ix,t) in enumerate(times[1:end-1])
    xs[ix+1] = xs[ix]+dt*x_dot(xs[ix],ys[ix],z)
    ys[ix+1] = ys[ix]+dt*y_dot(xs[ix],ys[ix],a,b)
end
scatter(x_0,y_0)
plot(xs,ys)
```

Here we see the stabe limit cycle
