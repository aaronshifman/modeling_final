```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
```

```python
def dV(V,W,alpha,z):
    return V*(alpha+V)*(1-V)-W+z
def dW(V,W,beta,c):
    return beta*V - c*W
```

```python
alpha = 0.01;
beta = 0.5;
c = 0.01;
z = 0.5;
dT = 0.005 #ms
time = 100 #ms

times = np.linspace(0,time,time/dT)

V= np.zeros(len(times))
W = np.zeros(len(times))
```

```python
for ix,t in enumerate(times[0:-1]):
    V[ix+1] = V[ix]+(dT*dV(V[ix],W[ix],alpha,z))
    W[ix+1] = W[ix]+(dT*dW(V[ix],W[ix],beta,c))
    
plt.plot(times,V)
plt.plot(times,W)
plt.legend(['V','W'])
```
